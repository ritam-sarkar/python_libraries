```python
import numpy as np
```

```python
python_list = [1,2,3]
print(python_list, type(python_list))
arr1 = np.array([1,2,3])
print(arr1, type(arr1))
```

#### Vectorized operation no lopp needed

```python
# print with lambda
print("add 10 with each x on python list",[x+10 for x in python_list])
print("vectorized add",arr1+10)
print("add in list with numpy",np.add(python_list,10))
```

Other methods of numpy

```python
print('print zeroes', np.zeros(3))
print('print ones for 3X2 dimension', np.ones((3,2)))
print('arrange 0 to 10 with gap of 2', np.arange(0,10,2))
print('random value between 2 to 3', np.random.rand(2,3))
```

Type promotion




int → float → complex → object

```python
arr_int = np.array([1, 2, 3])
arr_float = np.array([1.0, 2.0, 3.0])
arr_mixed = np.array([1, 2.5, 3])  # Mixed types → promoted to float
arr_mixed2 = np.array([1, "hello", 3])
print('promoted to float ',arr_mixed)
print('promoted to string object ',arr_mixed2)
```
